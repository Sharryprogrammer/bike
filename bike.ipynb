```python
da['Education']=pd.Categorical(da['Education'])
```

```python
da['Occupation']=pd.Categorical(da['Occupation'])
```

```python
da['Commute Distance']=pd.Categorical(da['Commute Distance'])
da['Region']=pd.Categorical(da['Region'])
da['Gender']=pd.Categorical(da['Gender'])
da['Home Owner']=pd.Categorical(da['Home Owner'])
```

```python
da['Marital Status']=pd.Categorical(da['Marital Status'])
```

```python
da=pd.get_dummies(da,drop_first=True)
da
```

```python
y=da["Purchased Bike"]
# X=da["Age","Region","Commute Distance","Cars","Home Owner","Occupation","Education","Children","Income","Gender","Marital Status"]
da=da.drop("ID",axis=1)#vertical
X=da.drop("Purchased Bike",axis=1)
da
```

```python
# from sklearn.ensemble import RandomForestRegressor
# regressor=RandomForestRegressor(n_estimators=150,random_state=0)
# regressor.fit(X,y)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators=100,random_state=0)
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
# 1-regressor.score(X_test, y_test)
```
